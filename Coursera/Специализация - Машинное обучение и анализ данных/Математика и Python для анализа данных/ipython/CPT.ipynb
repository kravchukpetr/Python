```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

* Берем **экспоненциальное распределение** с параметром лямбда равным единице
* Генерируем выборку объемом 1000 из  экспоненциального распределения

```python
exp_rv = sts.expon(0,1)
exp_sample = exp_rv.rvs(1000)
```

* Построим нормированную гистограмму выборки из экспоненциального распределения и на ней отобразим теоретическую плотность распределения случайной величины

```python
plt.hist(exp_sample, normed=True, bins=20,label=u'Гистограма')
plt.ylabel(u'кол-во элементов выборки')
plt.xlabel('$x$')

x = np.linspace(0,10,100)
pdf = exp_rv.pdf(x)
plt.plot(x, pdf, label=u'Теор. плотн. распр.')
plt.legend()
```

```python
# Сгенерируем 1000 выборок для каждого из объемов: 5,10,50,500 и расчитаем выборочные средние

sample_avg = {5:[],10:[],50:[],500:[]}

for n in [5,10,50,500]:
    i = 1
    while i <= 1000:
        sample = exp_rv.rvs(n)
        sample_avg[n].append(sample.mean())
        i = i + 1
```

```python
# Построим гистограмму распределения выборочных средних для n = 5

plt.hist(sample_avg[5], normed=True, bins=20, label=u'Гистограмма')
plt.ylabel(u'кол-во элементов выборки')
plt.xlabel('$x$')

norm_rv = sts.norm(1, (1.0/5) ** 0.5)
x = np.linspace(-2,3,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Теор. плотн. распр.')
plt.legend(loc='upper left')
```

```python
# Построим гистограмму распределения выборочных средних для n = 10

plt.hist(sample_avg[10], normed=True, bins=20, label=u'Гистограмма')
plt.ylabel(u'кол-во элементов выборки')
plt.xlabel('$x$')

norm_rv = sts.norm(1, (1.0/10) ** 0.5)
x = np.linspace(-2,3,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Теор. плотн. распр.')
plt.legend(loc='upper left')
```

```python
# Построим гистограмму распределения выборочных средних для n = 50

plt.hist(sample_avg[50], normed=True, bins=20, label=u'Гистограмма')
plt.ylabel(u'кол-во элементов выборки')
plt.xlabel('$x$')

norm_rv = sts.norm(1, (1.0/50) ** 0.5)
x = np.linspace(-2,3,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Теор. плотн. распр.')
plt.legend(loc='upper left')
```

```python
# Построим гистограмму распределения выборочных средних для n = 500

plt.hist(sample_avg[500], normed=True, bins=20, label=u'Гистограмма')
plt.ylabel(u'кол-во элементов выборки')
plt.xlabel('$x$')

norm_rv = sts.norm(1, (1.0/500) ** 0.5)
x = np.linspace(0.7,1.3,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Теор. плотн. распр.')
plt.legend(loc='upper left')
```

### Выводы

По итогам проведенных исследований стало явно видно, что точность аппроксимации увеличивается, при увеличении *n*, так, в приведенном примере, при *n = 5* аппроксимация уже довольно-таки неплоха, при *n = 10* гистограмма выборочных средних почти точно описывается нормальным распределением, та же тенденция сохраняется при дальнейшем увеличении *n*
